```python
ax = sb.countplot(data = df, x = 'Churn')

for bars in ax.containers : 
    ax.bar_label(bars)
mt.title("Count of customers by Churn")
    
#ax.bar_label(ax.containers[0]) ...for values in count
```

```python
mt.figure(figsize = (3,4))
gb = df.groupby('Churn').agg({'Churn': 'count'})
mt.pie(gb['Churn'], labels = gb.index, autopct = '%1.2f%%')
mt.title("Percentage of customers by Churn", fontsize = 10)
```

**From the given pie chart we can conclude that 26.54% of out customers have churned out**

```python
mt.figure(figsize = (5,4))
ax= sb.countplot(data = df, x = 'gender' ,hue = 'Churn')
ax.bar_label(ax.containers[0])
mt.title("Churn by Gender")
```

```python
mt.figure(figsize = (5,4))
sb.countplot(data = df, x = 'SeniorCitizen' ,hue = 'Churn')

mt.title("Churn by SeniorCitizenr")
```

```python
df_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack().fillna(0)

# Calculate the percentage of each 'Churn' within each 'SeniorCitizen'
df_percent = df_counts.div(df_counts.sum(axis=1), axis=0) * 100

# Plot the stacked bar chart
ax = df_percent.plot(kind='bar', stacked=True, figsize=(5, 4), color=['red', 'green'])

# Set the title and labels
ax.set_title("Churn by SeniorCitizen (Percentage)")
ax.set_ylabel("Percentage (%)")
ax.set_xlabel("Senior Citizen Status")

# Display the percentage on top of the bars
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x() + width / 2
    y = p.get_y() + height / 2
    ax.annotate(f'{height:.1f}%', (x, y), ha='center', va='center', fontsize=10, color='white')

# Show the plot
mt.tight_layout()
mt.show()
```

```python
mt.figure(figsize= (15,4))
sb.histplot(x = 'tenure', data = df, bins = 72, hue = 'Churn')
```

```python
selected_columns = ['PhoneService', 'MultipleLines', 'InternetService',
                    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                    'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots (rows, columns)
n_cols = 3  # Set the number of columns for the subplot grid
n_rows = (len(selected_columns) + n_cols - 1) // n_cols  # Calculate number of rows

# Create figure and subplots
fig, axes = mt.subplots(n_rows, n_cols, figsize=(15, 12))
axes = axes.flatten()  # Flatten the 2D array of axes for easier iteration

# Loop through the selected columns and create a countplot on each axis
for i, col in enumerate(selected_columns):
    sb.countplot(data=df, x=col, ax=axes[i], palette='Set2', hue ='Churn')  # Set a color palette
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove any empty subplots if the total number of subplots doesn't perfectly match
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust the layout to make sure plots are well-spaced
mt.tight_layout()
mt.show()
```

***The visualizations display the relationship between churn (Yes/No) and various telecom services like Phone Service, Internet Service, and add-on features such as Online Security, Tech Support, and Streaming. Across all plots, customers with No or inactive services (e.g., no Internet, no Online Security) show a higher churn rate (orange bars). Fiber optic Internet customers also exhibit higher churn compared to DSL users. Additionally, features like Tech Support, StreamingTV, and OnlineBackup show a notable reduction in churn for customers who actively use these services.***
